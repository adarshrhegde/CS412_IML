```python
import pandas as pd
```

```python
''' Read Orders dataset'''
orders_df = pd.read_csv('data/olist_orders_dataset.csv')
orders_df = orders_df[['order_id', 'customer_id', 'order_purchase_timestamp']]
orders_df.head()
```

```python
''' Read order items dataset '''
order_items_df = pd.read_csv('data/olist_order_items_dataset.csv')
order_items_df = order_items_df[['order_id', 'product_id', 'price']]
order_items_df.head()
```

```python
''' Merge orders and order items to get product id and price for that order '''
orders_df = orders_df.merge(order_items_df, how='left', on='order_id')
orders_df.head()
```

```python
''' Read products dataset '''
products_df = pd.read_csv('data/olist_products_dataset.csv')
products_df = products_df[['product_id', 'product_category_name']]
products_df.head()
```

```python
''' Merge orders and products datasets to get product category name in foreign lang '''
orders_df = orders_df.merge(products_df, how='left', on='product_id')
orders_df.head()
```

```python
''' Read product category dataset'''
product_cat_name_df = pd.read_csv('data/product_category_name_translation.csv')
product_cat_name_df.head()
```

```python
''' Merge orders and product cat name datasets to get product category name in eng '''
orders_df = orders_df.merge(product_cat_name_df, how='left', on='product_category_name')
orders_df.head()
```

```python
''' Read customers dataset '''
customers_df = pd.read_csv('data/olist_customers_dataset.csv')
customers_df = customers_df[['customer_id', 'customer_unique_id', 'customer_state']]
customers_df.head()
```

```python
''' Merge orders and customers datasets to get customer state '''
orders_df = orders_df.merge(customers_df, how='left', on='customer_id')
orders_df.head()
```

```python
orders_df.to_csv('test.csv')
```
