```python
import pandas as pd
import datetime
```

```python
#Read customer's dataset
olist_orders_df = pd.read_csv('data/olist_orders_dataset.csv')
olist_orders_df = olist_orders_df[['order_id','customer_id', 'order_purchase_timestamp']]
olist_items_df = pd.read_csv('data/olist_order_items_dataset.csv')
olist_items_df = olist_items_df[['order_id', 'product_id', 'price']]
olist_products_df = pd.read_csv('data/olist_products_dataset.csv')
olist_products_df.head()
olist_orders_df = olist_orders_df.merge(olist_items_df, how = "left", on = "order_id")
olist_products_df.head()
```

```python
olist_products_df = olist_products_df[['product_id', 'product_category_name']]
olist_products_df.head()
olist_orders_df = olist_orders_df.merge(olist_products_df, how = 'left', on = 'product_id')
olist_orders_df.head()
```

```python
olist_english_products = pd.read_csv('data/product_category_name_translation.csv')
olist_orders_df = olist_orders_df.merge(olist_english_products, how = "left", on = 'product_category_name')
olist_orders_df.head()
```

```python
olist_customer_df = pd.read_csv('data/olist_customers_dataset.csv')
olist_customer_df = olist_customer_df[['customer_id','customer_unique_id','customer_city','customer_state']]
olist_customer_df.head()
olist_orders_df = olist_orders_df.merge(olist_customer_df, how = 'left', on = 'customer_id')
```

```python
olist_orders_df['order_purchase_timestamp'] = pd.to_datetime(olist_orders_df['order_purchase_timestamp'], format="%Y-%m-%d %H:%M:%S")
```

```python
olist_orders_df['order_purchase_month'] = olist_orders_df['order_purchase_timestamp'].apply(lambda x: x.month)
```

```python
olist_month_split = olist_orders_df.groupby(['order_purchase_month'], as_index=False).groups
```
