```python
import pandas as pd
import numpy as np
```

```python
''' Read Orders dataset'''
orders_df = pd.read_csv('data/olist_orders_dataset.csv')
# orders_df = orders_df[['order_id', 'customer_id', 'order_purchase_timestamp']]
print(orders_df.shape)
orders_df.head()
```

```python
''' Read order items dataset '''
order_items_df = pd.read_csv('data/olist_order_items_dataset.csv')
order_items_df = order_items_df[['order_id', 'product_id', 'price']]
print(order_items_df.shape)
order_items_df.head()
```

```python
''' Merge orders and order items to get product id and price for that order '''
orders_df = orders_df.merge(order_items_df, how='right', on='order_id')
print(orders_df.shape)
orders_df.to_csv('test.csv', index=False)
orders_df.head()
```

```python
''' Read products dataset '''
products_df = pd.read_csv('data/olist_products_dataset.csv')
products_df = products_df[['product_id', 'product_category_name']]
print(products_df.shape)
products_df.head()
```

```python
''' Merge orders and products datasets to get product category name in foreign lang '''
orders_df = orders_df.merge(products_df, how='left', on='product_id')
print(orders_df.shape)
orders_df.to_csv('test.csv', index=False)
orders_df.head()
```

```python
''' Removing rows which don't have product category name '''
orders_df = orders_df[pd.notnull(orders_df['product_category_name'])]
print(orders_df.shape)
orders_df.to_csv('test.csv', index=False)
orders_df.head()
```

```python
''' Read product category dataset'''
product_cat_name_df = pd.read_csv('data/product_category_name_translation.csv')
print(product_cat_name_df.shape)
product_cat_name_df.head()
```

```python
''' Merge orders and product cat name datasets to get product category name in eng '''
orders_df = orders_df.merge(product_cat_name_df, how='left', on='product_category_name')
print(orders_df.shape)
orders_df.to_csv('test.csv', index=False)
orders_df.head()
```

```python
''' Read customers dataset '''
customers_df = pd.read_csv('data/olist_customers_dataset.csv')
customers_df = customers_df[['customer_id', 'customer_unique_id', 'customer_state']]
print(customers_df.shape)
customers_df.head()
```

```python
''' Merge orders and customers datasets to get customer state '''
orders_df = orders_df.merge(customers_df, how='left', on='customer_id')
print(orders_df.shape)
orders_df.to_csv('test.csv', index=False)
orders_df.head()
```

```python
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'], format="%Y-%m-%d %H:%M:%S")
orders_df.head()
```

```python
''' Adding specific month and year column '''
orders_df['order_purchase_year_month'] = orders_df['order_purchase_timestamp'].apply(lambda x: 100*x.year + x.month)
orders_df.head()
```

```python
final_df = orders_df.drop(['customer_id', 'order_purchase_timestamp', 'product_id', 'product_category_name', 'customer_unique_id'], axis=1)
orders_df.to_csv('final.csv', index=False)
final_df.head()
```

```python
final_df.shape
```

```python
final_df = final_df.groupby(['product_category_name_english', 'customer_state', 'order_purchase_year_month'], as_index=False).agg({
    'order_id': 'count',
    'price': 'sum'
})
print(final_df.shape)
final_df.head()
```

```python
final_df = final_df.sort_values(['product_category_name_english', 'order_purchase_year_month', 'price'], ascending=[1, 1, 0])
final_df.head()
```

```python
final_df = final_df.groupby(['product_category_name_english', 'order_purchase_year_month']).first().reset_index().rename(columns={'order_id': 'order_count'})
final_df.to_csv('clustering_data.csv', index=False)
final_df
```

```python
final_df['order_purchase_year_month'] = final_df['order_purchase_year_month'].apply(lambda x: x % 100)
final_df.to_csv('clustering_data.csv', index=False)
final_df.head()
```

```python
final_one_hot_df = pd.get_dummies(final_df, columns = ['product_category_name_english', 'order_purchase_year_month', 'customer_state'])
final_one_hot_df.to_csv('clustering_data_one_hot.csv', index=False)
final_one_hot_df.head()
```

```python
x_cols = ['order_count', 'price']
y_cols = []

for elem in final_one_hot_df.columns.values:
    if elem.startswith('product_category_name') or elem.startswith('order_purchase_year_month'):
        x_cols.append(elem)
    elif elem.startswith('customer_state'):
        y_cols.append(elem)
```
